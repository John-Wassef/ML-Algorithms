```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv("Second Wizard War DeadorAliveUpdated.csv")
df.info()
```

```python
df=df.drop(['NAME'],axis=1)
df
```

```python
df=df.dropna()
df = df.loc[df.index]
print(df.isnull().sum())
```

```python
df['ALIVE?']=df['ALIVE?'].replace({'Yes': 1, 'No': 0})
df = pd.get_dummies(df, columns=['BLOOD STATUS', 'Species', 'GENDER', 'HOUSE', 'HAIR COLOR', 'BOGGART', 'PATRONUS', 'Nationality',])
```

```python
df = df.astype(int)
X=df.drop(['ALIVE?'],axis=1)
Y=df['ALIVE?']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=175)

print('X train size:',len(x_train))
print('Y train size:',len(y_train))
print('X test size:',len(x_test))
print('X test size:',len(y_test))
```

Logistic Model

```python
logistic_model=LogisticRegression()
logistic_model.fit(x_train,y_train)
logistic_pred=logistic_model.predict(x_test)
logistic_accuracy=logistic_model.score(x_test,y_test)
logistic_f1score=f1_score(y_test,logistic_pred)
print("Logistic F1 Score:",logistic_f1score)
print("Logistic accuracy:",logistic_accuracy)
```

Decision Tree Model

```python
decisionTree_model=DecisionTreeClassifier()
decisionTree_model.fit(x_train,y_train)
decisionTree_pred=decisionTree_model.predict(x_test)
decisionTree_accuracy=decisionTree_model.score(x_test,y_test)
decisionTree_f1score=f1_score(y_test,decisionTree_pred)
print("Decision Tree F1 Score:",decisionTree_f1score)
print("Decision Tree accuracy:",decisionTree_accuracy)
```

SVM Model

```python
SVM_model=SVC()
SVM_model.fit(x_train,y_train)
SVM_pred=SVM_model.predict(x_test)
SVM_accuracy=SVM_model.score(x_test,y_test)
SVM_f1score=f1_score(y_test,SVM_pred)
print("SVM F1 Score:",SVM_f1score)
print("SVM accuracy:",SVM_accuracy)
```

Random Forest Model

```python
randomForest_model=RandomForestClassifier()
randomForest_model.fit(x_train,y_train)
randomForest_pred=randomForest_model.predict(x_test)
randomForest_accuracy=randomForest_model.score(x_test,y_test)
randomForest_f1score=f1_score(y_test,randomForest_pred)
print("Random Forest F1 Score:",randomForest_f1score)
print("Random Forest accuracy:",randomForest_accuracy)
```

Naive Bayes Model

```python
naive_bayes_model = GaussianNB()
naive_bayes_model.fit(x_train, y_train)
naive_bayes_pred=naive_bayes_model.predict(x_test)
naive_bayes_accuracy=naive_bayes_model.score(x_test,y_test)
naive_bayes_f1score=f1_score(y_test,naive_bayes_pred)
print("Naive Bayes F1 Score:",naive_bayes_f1score)
print("Naive Bayes accuracy:",naive_bayes_accuracy)
```

KNN Model

```python
knn_model = KNeighborsClassifier(n_neighbors=4)
knn_model.fit(x_train, y_train)
knn_pred=knn_model.predict(x_test)
Knn_accuracy=knn_model.score(x_test,y_test)
knn_f1score=f1_score(y_test,knn_pred)
print("KNN F1 Score:",knn_f1score)
print("KNN accuracy:",Knn_accuracy)
```

```python
results = {
    
    'Model': ['Logistic Model', 'Decision Tree Model', 'SVM Model', 'Random Forest Model', 'Naive Bayes Model', 'KNN Model'],
    'F1 Score': [logistic_f1score, decisionTree_f1score, SVM_f1score, randomForest_f1score, naive_bayes_f1score, knn_f1score],
    'Accuracy Score': [logistic_accuracy, decisionTree_accuracy, SVM_accuracy, randomForest_accuracy, naive_bayes_accuracy, Knn_accuracy ],
}

summary_df = pd.DataFrame(results)
summary_df
```
